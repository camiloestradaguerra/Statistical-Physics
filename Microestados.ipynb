# **Sistema de Espines no interactuantes**

```python
import scipy.special
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# Entrada de datos
n_spins = 10
magns_array = np.arange(-n_spins, n_spins + 1, 1)

def microstates(magnetization):
    if n_spins >= magnetization:
        n_up = n_spins + magnetization
        n_down = n_spins - magnetization
        microstates_number = scipy.special.binom(n_spins, 0.5*(n_spins +magnetization))
    
    return int(microstates_number)
```

```python
microstates_array = np.array([])
for i in range(0, 2*n_spins + 1):
    microstates_array = np.append(microstates_array, microstates(magns_array[i]))
```

```python
#plt.rcParams["figure.figsize"] = (10,10)
plt.figure(figsize=(12,8))
plt.plot(magns_array, microstates_array)
plt.scatter(magns_array, microstates_array, color = 'red')
plt.xticks(np.arange(magns_array.min(), magns_array.max()+1, step=1), size=15)
plt.yticks(microstates_array, size=15)
plt.xlabel("Magnetización", size=15)
plt.ylabel('Número de Microestados', size=15)
plt.show()
```
